```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re
```

```python
#for reading the csv file
data = pd.read_csv("car_price_performance.csv")
data
```

```python
brand_avg_prices = data.groupby("Brand")["Price"].mean().sort_values()
brand_avg_prices
```

```python
#Encoding the brand
brand = {}
count=1
for i in brand_avg_prices.index:
    brand[i]=count
    count+=1
brand
```

```python
val=[]
for j in data["Brand"]:
    y=brand[j]
    val.append(y)
data['Encoded Brand']= val
data
```

```python
scaledBrand=[]
min= data['Encoded Brand'].min()
max= data['Encoded Brand'].max()
for x in data['Encoded Brand']:
    scaled_value = (x-min)/(max-min)
    scaledBrand.append(scaled_value)
data['Scaled Encoded Brand'] = scaledBrand
data
```

```python
scaledHorsepower=[]
min= data['Horsepower'].min()
max= data['Horsepower'].max()
for x in data['Horsepower']:
    scaled_value = (x-min)/(max-min)
    scaledHorsepower.append(scaled_value)
data['Scaled Encoded Horsepower'] = scaledHorsepower
data
```

```python
scaledFuelEfficiency=[]
min= data['Fuel Efficiency'].min()
max= data['Fuel Efficiency'].max()
for x in data['Fuel Efficiency']:
    scaled_value = (x-min)/(max-min)
    scaledFuelEfficiency.append(scaled_value)
data['Scaled Encoded Fuel Efficiency'] = scaledFuelEfficiency
data
```

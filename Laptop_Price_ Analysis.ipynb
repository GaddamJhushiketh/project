```python
# Calculate average prices by brand
brand_avg_prices = data.groupby("Company")["Price"].mean().sort_values(ascending=False)
brand_avg_prices
```

```python
#for ploting the graph based on the average price of laptop based on companyes
sns.barplot(x=brand_avg_prices.index, y=brand_avg_prices.values)

plt.title("Average Laptop Prices by Brand", fontsize=16)
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=90)
plt.show()
```

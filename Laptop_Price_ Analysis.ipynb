```python
def get_storage_type(row):
    memory = str(row).lower()
    total_weighted_gb = 0
    #memory_type=['ssd','flash','hybrid','hdd']
    #print(memory)
    if 'ssd' in memory and ('flash' not in memory and 'hybrid' not in memory and 'hdd' not in memory):
        stype = 'SSD'
    elif 'flash' in memory and ('ssd' not in memory and 'hybrid' not in memory and 'hdd' not in memory):
        stype = 'Flash Storage'
    elif 'hybrid' in memory and ('flash' not in memory and 'ssd' not in memory and 'hdd' not in memory):
        stype = 'Hybrid'
    elif 'hdd' in memory and ('flash' not in memory and 'hybrid' not in memory and 'ssd' not in memory):
        stype = 'HDD'
    elif 'hdd' in memory and 'ssd' in memory  and ('flash' not in memory and 'hybrid' not in memory):
        stype = 'HDD+SSD'
    elif 'flash' in memory and 'hybrid' in memory  and ('ssd' not in memory and 'hdd' not in memory):
        stype = 'Flash+Hybrid'
    elif 'hdd' in memory and 'flash' in memory  and ('ssd' not in memory and 'hybrid' not in memory):
        stype = 'HDD+Flash'
    elif 'ssd' in memory and 'hybrid' in memory  and ('hdd' not in memory and 'flash' not in memory):
        stype = 'SSD+Hybrid'
    else:
      stype='Others'
    return stype
data['Storage_Type'] = data['Memory'].apply(get_storage_type)
```

```python
data['Storage_Type'].value_counts()
```

```python
ssd_data = data[data['Storage_Type'].str.contains('SSD', case=False, na=False)]
hdd_data = data[data['Storage_Type'].str.contains('HDD', case=False, na=False)]
```

```python
def get_memory(row):
  if "+" in row:
    return 0
  memory_str=row.lower().strip()
  memory = re.search(r'(\d+(?:\.\d+)?)\s*(gb|tb)', memory_str)
  size = float(memory.group(1))
  #print(size)
  unit = memory.group(2)
  #print(unit)
  if 'tb' in unit:
    size = size*1024
  return size
```

```python
ssd_data['Memory_Capacity']=ssd_data['Memory'].apply(get_memory)
hdd_data['Memory_Capacity']=hdd_data['Memory'].apply(get_memory)
```

```python
ssd_df=ssd_data[ssd_data['Memory_Capacity']!=0]
hdd_df=hdd_data[hdd_data['Memory_Capacity']!=0]
```

```python
ssd_data_sorted = ssd_data.sort_values(by='Memory_Capacity')
hdd_data_sorted = hdd_data.sort_values(by='Memory_Capacity')
```

```python
plt.figure(figsize=(10, 6))
# Plotting SSD data
plt.plot(ssd_data_sorted['Memory_Capacity'], ssd_data_sorted['Price'], label='SSD', color='yellow')

# Plotting HDD data
plt.plot(hdd_data_sorted['Memory_Capacity'], hdd_data_sorted['Price'], label='HDD', color='blue')

plt.title('Price vs Memory Capacity: SSD vs HDD')
plt.xlabel('Memory Capacity (in GB)')
plt.ylabel('Price')

plt.show()
```

```python
def weighted_memory(row):
    # if pd.isna(row):
    #     return 0
    mem_string=row
    # Define weights
    weights = {
        'SSD': 1.0,
        'Flash Storage': 0.8,
        'Hybrid': 0.6,
        'HDD': 0.5
    }

    total = 0
    parts = mem_string.split('+') if '+' in mem_string else [mem_string]

    for i in parts:
        part = i.lower().strip()
        match_string = re.search(r'(\d+(?:\.\d+)?)\s*(gb|tb)', part)
        size = float(match_string.group(1))
        unit = match_string.group(2)
        weight = 0
        if 'tb' in part:
          size = size*1024
        if 'flash' in part:
            weight = weights['Flash Storage']
        elif 'ssd' in part:
            weight = weights['SSD']
        elif 'hdd' in part:
            weight = weights['HDD']
        elif 'hybrid' in part:
            weight = weights['Hybrid']
        #print(size)
        total += size * weight

    return total
```
